```python
# we can nest one list comprehension below another # you can write it like writing SQL subquery
# think about what u looking for and write it inner out

nested=[letters[1] for letters in [a+b for a in 'life' for b in 'study']] # note a+b was the previous combined list comprehension
print(nested)
```

```python
# better way to write it 
combined =[a+b for a in "life" for b in 'study']
non_nested=[letters[1] for letters in combined]
print(non_nested)
```

```python
words=['life','is','study']
word_length_dict={} # declare the dict

for word in words: 
    word_length_dict[word]=len(word) # regular loop but obey the dict structure
print(word_length_dict)
```

```python
# dict comprehension kick in 
words=['life','is','study']
word_length_dict2={word:len(word) for word in words} # note here like list [] ommit append; dict {} since comprehension within 
# disregard the dict structure;  
# : separate the key and value; after : you can put operation to generate the value + logic iterate over

print(word_length_dict2)
```

```python
# use dict to mapping two saperate sequence; eg 2 columns in pandas----zip
words=['life','is','study']
word_lengths=[4,2,5]
pairs=zip(words,word_lengths)

for item in pairs:
    print(item)
    
```

```python
# above tuple pair the key and elements; extracted by zip and ready to be come dicts
words=['life','is','study']
word_lengths=[4,2,5]
w_dict3={key:value for (key,value) in zip(words,word_lengths)}
#### note key:value -- it specify save the tuple as key & value pairs

print(w_dict3)
```
