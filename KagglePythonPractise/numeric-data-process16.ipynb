```python
import numpy as np
```

```python
normal_distribute=np.random.normal(size=1000)
normal_distribute[0:15]
```

```python
pd.DataFrame(normal_distribute).hist(figsize=(8,8),bins=30)
```

```python
# use lineregression the assumption is the distribution is normally distributed
```

```python
skewed=np.random.exponential(scale=2,size=1000)
skewed=pd.DataFrame(skewed)
skewed.hist(figsize=(8,8),bins=50)
```

```python
# above is right skewed -skew to the right
```

```python
# we can transform the data with sqrt to make it less skew---for ai study--using
# .apply
```

```python
sqt_transformed=skewed.apply(np.sqrt) # note here we use a apply function on a data frame
sqt_transformed
```

```python
sqt_transformed.hist(figsize=(
    8,8),bins=50) # sligthly better
```

```python
# log transform has to +1; eg. less than 1 log get negtaive; zero will not have value
```

```python
log_transformed=(skewed+1).apply(np.log)
log_transformed.hist(figsize=(8,8),bins=50);
```

```python
# sqrt in this case is better--reduce some obvious skewed
```
