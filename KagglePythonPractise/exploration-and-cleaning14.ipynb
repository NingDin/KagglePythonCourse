```python
# to replace the missing data
#1. replace null value with 0
#2. replace the null value with some cnetral value eg. mean / median # if it is normal distr; making sense to replace with mean/meidan
#3. impute some other value- self def calcualtion/rules
#4.split the data set into two parts: one set with where records have an age value and another set where ae is null--split sets
```

```python
# to see weather age is normal distributed by having hist
titanic_train.hist(column='Age',figsize=(9,6),bins=15)
```

```python
# we saw a concentration on lower age
# since it is kind skew we use median rather than mean
```

```python
new_age_var=np.where(titanic_train['Age'].isnull(),titanic_train['Age'].median(),titanic_train['Age']) # we do not to filter 
# titanic_train['Age'].isnull()==True; reason we have np where for true and false; ture assign median; false keep the same value origin
titanic_train['Age']=new_age_var
titanic_train['Age'].describe()
```
