```python
table1 = pd.DataFrame({"P_ID" : (1,2,3,4,5,6,7,8),
                     "gender" : ("male", "male", "female","female",
                                "female", "male", "female", "male"),
                     "height" : (71,73,64,64,66,69,62,72),
                     "weight" : (175,225,130,125,165,160,115,250)})

table1
```

```python
table2 = pd.DataFrame({"P_ID" : (1, 2, 4, 5, 7, 8, 9, 10),
                     "sex" : ("male", "male", "female","female",
                            "female", "male", "male", "female"),
                     "visits" : (1,2,4,12,2,2,1,1),
                     "checkup" : (1,1,1,1,1,1,0,0),
                     "follow_up" : (0,0,1,2,0,0,0,0),
                     "illness" : (0,0,2,7,1,1,0,0),
                     "surgery" : (0,0,0,2,0,0,0,0),
                     "ER" : ( 0,1,0,0,0,0,1,1) } ) 

table2
```

```python
# p_id is the common col
```

```python
pd.merge(table1,table2,how='inner',on='P_ID')
```

```python
# left join take all from first table and match second as much as it can ; if not still keep the first table and return NaN
```

```python
pd.merge(table1,table2,how='left',on='P_ID')
```

```python
# right join is flip of left
pd.merge(table1,table2,how='right',on='P_ID')
```

```python
# keep all - outer
pd.merge(table1,table2,how='outer',on='P_ID')
```

```python
# note gender and sex same thing we can outter join on multiple common columns
```

```python
# first we need to rename to unify column name
table1.rename(columns={'gender':'sex'},inplace=True)
```

```python
pd.merge(table1,table2,how='outer',on=['P_ID','sex']) # note if merge on multiple col use [] as container to hold diff common/matching cols name
```
